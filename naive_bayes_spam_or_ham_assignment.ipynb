```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv('combined_data.csv')
df.head(10)
```

```python
Vectorizer=TfidfVectorizer()
X=df['text'].values
Y=df['label'].values
print(X)
print(Y)
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.2)
```

```python
#X_transform=Vectorizer.fit_transform(X)
vectorized_x_train=Vectorizer.fit_transform(x_train)
vectorized_x_test=Vectorizer.transform(x_test)
```

```python
model=MultinomialNB()
model
```

```python
model.fit(vectorized_x_train,y_train)
```

```python
y_pred=model.predict(vectorized_x_test)
from sklearn.metrics import accuracy_score,recall_score, f1_score,confusion_matrix
acc=accuracy_score(y_pred,y_test)
print("accuracy:",acc)
rec=recall_score(y_pred,y_test)
print("recall:",rec)
f=f1_score(y_pred,y_test)
print("f1_score:",f)
user_input=input("enter:")
a=[user_input]
a_transform=Vectorizer.transform(a)
b=model.predict(a_transform)
print(b)
og={1:'ham',2:'spam'}
print(og[b[0]])
```

```python
import seaborn as sns
confusion_matrix=confusion_matrix(y_pred,y_test)
sns.heatmap(confusion_matrix,cmap='coolwarm',annot=True,xticklabels=['Ham', 'Spam'], yticklabels=['Ham','Spam'])
import matplotlib.pyplot as plt
plt.xlabel("predicted")
plt.ylabel("actual")
plt.title("confusion matrix")
plt.show()
```
